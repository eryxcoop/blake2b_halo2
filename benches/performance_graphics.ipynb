```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
dataframes = []
for optimization in ["A", "B", "C"]:
    for blocks in [1,5,10,15,20]:
        df = pd.read_csv(f"../../target/criterion/optimization_comparison/{optimization}/{blocks}/new/raw.csv")        
        df["measure"] = df["sample_measured_value"]/df['iteration_count']
        df = df.rename(columns={"iteration_count": "iteration", "function": "optimization", "throughput_num": "blocks"})
        dataframes.append(df[["optimization", "blocks", "iteration", "measure"]])
all_data = pd.concat(dataframes)
print(all_data)
```

```python
sns.lineplot(x='blocks', y='measure', hue="optimization", data=all_data)
# sns.boxplot(x='blocks', y='measure', hue="optimization", data=all_data)
```
